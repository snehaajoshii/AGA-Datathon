```python
import pandas as pd 
```

```python
df = pd.read_excel('/Users/snehajoshi/Downloads/MA Merged Data.xlsx' , sheet_name = 'T13&T14')
```

```python
df
```

```python
df = df[df['Sector'] == 'All Reported Sectors']
```

```python
df
```

```python
df = df.drop('Sector', axis=1)
```

```python
df
```

```python
df.set_index('State', inplace=True)

# Multiply all values by 100 to convert them to percentages
df = df * 100
```

```python
df.reset_index(inplace=True)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8 ))
for year in [2008, 2009,2010,2011,2012,2013,2014,2015]:
    plt.plot(df['State'], df[year], label=year)

plt.title('Graph')
plt.xlabel('State')
plt.ylabel('rate')
plt.legend(title='Year')
plt.grid(True)
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

```python
df.to_excel("gradratebystate.xlsx", index=False) 
```

```python
df = df.melt(id_vars=["State"], var_name="Year", value_name="Value")

# Sort the DataFrame by State and Year
df = df.sort_values(by=["State", "Year"]).reset_index(drop=True)

# Display the reshaped DataFrame
print(df)
```
