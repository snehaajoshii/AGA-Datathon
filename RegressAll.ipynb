Step 2: Look at the regression between unemployed people per year and graduation rate

```python
import numpy as np
import scipy.stats as stats

x = df['Unemployment by state per year']
y = df['Graduation rate']

# Perform linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

# Print the regression values
print(f"Slope: {slope:.6f}")
print(f"Intercept: {intercept:.6f}")
print(f"R-squared: {r_value**2:.6f}")
print(f"P-value: {p_value:.6f}")
print(f"Standard Error: {std_err:.6f}")
```

The above program shows the relationship between graduation rate and unemployed people per state throughout the years. It shows that they have  a negative relationship where the graduation rate is expected to decrease by approximately 0.00062% when the unemployed person per state increases by 1 person. With R-squared being 0.003736, we can say that 0.3% of the variation in graduation rate is explained by the variable unemployed people per state. With the p-value being 0.24473, which is greater than 0.05, we can say that this relationship shown by the regression is not statistically significant.

The program below creates a scatterplot using the seaborn library in python to create a joint plotof bar chart as well as line graph to visualize the relationship between unemployed people per state by year and graduation rate.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="darkgrid")
g = sns.jointplot(x=x, y=y, kind="reg", truncate=False, color="m", height=7)

# Customize the labels for x and y axes
g.set_axis_labels("Unemployment by state per year", "Graduation Rate")

# Show the plot
plt.show()
```
