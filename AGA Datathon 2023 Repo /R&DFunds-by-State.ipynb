```python
import pandas as pd 
```

```python
df = pd.read_excel('/Users/snehajoshi/Downloads/Federal R&D Obligations (1)..xlsx')
```

```python
df.set_index('State', inplace=True) #setting State(str) as index temporary to make all the other columns same object
row_averages = df.mean(axis=1) #taking the avg of every row
df = df.apply(lambda row: row.fillna(row_averages)) #filling the empty cell with the avg for accuracy
```

```python
df= df.round() #round the data to the nearest value 
df= df.astype(int) #setting to int
```

```python
#creating new data frame for visualization 
new_data = df.set_index('State')[[2003, 2009, 2019]]
# Reset the index of the new DataFrame
new_data.reset_index(inplace=True)

# Print the resulting DataFrame
new_data
```

```python
#plot each year on top of one another as a line chart for comparison
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
for year in [2003, 2009, 2019]:
    plt.plot(new_data['State'], new_data[year], label=year)

plt.title('R&D Obligations by State per individual')
plt.xlabel('State')
plt.ylabel('R&D Funds')
plt.legend(title='Year')
plt.grid(True)
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```
