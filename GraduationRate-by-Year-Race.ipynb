```python
import pandas as pd
```

```python
df= pd.read_excel('/Users/snehajoshi/Downloads/Graduation visualization.xlsx')
```

```python
df.columns
```

```python
df= df.drop('Completion at Same Institution (%)', axis=1)
df= df.drop('Completion at Different 4YR (%)', axis=1)
df= df.drop('Still Enrolled (Anywhere, %)', axis=1)
df= df.drop('No Longer Enrolled (Anywhere, %)', axis=1)
df= df.drop('Completion at Different 2YR (%)', axis=1)
df
```

```python

# Create a DataFrame from the data
df = pd.DataFrame(df)

# Pivot the DataFrame to have the years as columns and race/ethnicity as index
df_pivot = df.pivot(index='Race/Ethnicity', columns='Cohort Year', values='Total Completion (%)')

# Display the modified DataFrame
df_pivot
```

```python
df_pivot.columns.name = 'Year'
```

```python
import matplotlib.pyplot as plt

# Set the figure size
plt.figure(figsize=(10, 6))

# Iterate through the race/ethnicity categories
for race in df_pivot.index:
    # Plot the data for each race/ethnicity
    plt.plot(df_pivot.columns, df_pivot.loc[race], marker='o', label=race)

# Set labels and title
plt.title('Graduation Rate Throughout Year By Race')
plt.xlabel('Year')
plt.ylabel('Graduation Rate (Public Institutions)')

# Add a legend and grid
plt.legend(title='Race/Ethnicity')

# Rotate x-axis labels for better readability
plt.xticks(df_pivot.columns, rotation=90)

# Show the plot
plt.show()
```
